# Udemy Course Project

This dataset contains various information about Udemy courses. It includes detailed data such as the title, price, number of subscribers, number of reviews, number of lectures, content duration, and published date of each course. It also shows whether the course is paid, its level (beginner, intermediate, expert), and the subject area it belongs to (e.g., business, web development, musical instruments, etc.). This dataset is a great resource for educational analysis, pricing strategies, and examining the impact of educational content on user demand.

```python
labels = ['Low','Medium','High']
bins = [0,50,150,float('inf')]
df['price_category'] = pd.cut(df['price'],bins=bins,labels=labels,right=False)
```

```python
price_dist = df.price_category.value_counts()
price_dist
```

```python
plt.pie(x= price_dist.values,
        labels= price_dist.index,
        autopct="%1.1f%%",
        explode= [0, 0.03, 0.1],
        colors= ["#76b7b2", "#f28e2b", "#e15759"],
        shadow= True,
        startangle= 180
       )

plt.show()
```

- This chart shows the distribution of prices. Most of the courses are priced upton 25 USD, with a notable presence of free and low price courses.
- As the price increases, the number of courses decreases.
- However, a few courses priced at 200 USD indicates that high price courses also attract intrest. 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'price', y= 'num_subscribers',hue='price_category',data = df,color = 'purple')
plt.title('Price Vs Number of Subscribers')
plt.xlabel('Price')
plt.ylabel('Number Of Subscribers')
plt.show()
```

- Low price subscribers particularly free ones tends to attrack signoficantly more subscribers.
- This suggests that affortable courses has strong influence in students enrollment.
- High price courses shows that very few courses at this level,and this courses genrelly have fewer subscribers, though there are some exceptions. 

```python
df["revenue"] = df["price"] * df["num_subscribers"]

revenue_by_year = df.groupby('Published_Year')['revenue'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(revenue_by_year['Published_Year'], revenue_by_year['revenue'], marker='o', color='orange')

plt.title('Total Revenue by Year')
plt.ylabel('Total Revenue')

plt.show()
```

- This graph shows the total revenue by year. From 2011 to 2015, there is steady growth in revenue, reaching its peak in 2015. However, after 2015, there is a noticeable decline, and by 2017, revenue drops below the 2014 level. This suggests that after a period of rapid growth, the platform experienced a decline in revenue.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df, x= 'level',y='content_duration')
plt.title('Content Duration By Course Level')
plt.xlabel('Course Level')
plt.ylabel('Content Duration(hours)')
plt.show()
```

- All Level, Immediate Level and Beginner Level courses have similar median duration around 10 hours, with some outliers reaching upto hours.
- Expert level courses are shorter on average, with median around 5 to 7 hours, and few outliers reaching 30 hours. 

```python
top10_courses = df.sort_values(by= 'num_subscribers',ascending=False)[:10]

plt.figure(figsize=(10,6))
ax = sns.barplot(x='num_subscribers',y='course_title',data = top10_courses)

for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=9, color='black', padding=3)


plt.title('Top 10 Most Popular Courses by Subscribers', fontsize=16)
plt.xlabel('Number of Subscribers', fontsize=12)
plt.ylabel('Course Title', fontsize=12)
plt.yticks(fontsize= 8)

plt.show()
```

##### This bar chart presents the top 10 most popular Udemy courses based on the number of subscribers.
- 
Learn HTML5 Programming From Scratch is the most popular course, with over 250,000 subscribers- .
Other popular topics include entrepreneurship basics, web development, and HTML/CSS, indicating a strong interest in technical and business-related skill- s.
While technical courses dominate the list, it's notable that a non-technical course like Beginner Electric Guitar Lessons also ranks highly, showing diversity in learner interests.

```python
plt.figure(figsize=(10,6))
plt.scatter(df['num_lectures'],df['num_reviews'],color = 'green',s=100)
plt.xlabel('Number Of Lecture')
plt.ylabel('Number Of Reviews')
plt.title('Lecture Vs Reviews')
plt.show()
```

- This Graph shows the relation between the number of lecture and number of reviews.
- Courses with a fewer lectures, particularly between 0 to 100 tend to have more reviews.
- As the number of lectures increses their isen;t significant rise in the number of reviews.
- Courses with 300 lectures tend to receive fewer reviews, indicating that more content dosen't necessarily lead to more engagement in terms of reviews.

```python
correlation_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_df.corr(),annot=True)
```

##### This heatmap shows correlations between variables, where the intensity of the color indicates the strength of the relationship.
##### Revenue:
- Has a strong positive correlation with number of reviews (0.77), indicating that courses with more reviews tend to generate higher revenue.
- There is also a positive relationship with number of subscribers (0.56), showing that more subscribers contribute to higher revenue.
- Number of lectures and duration have weaker positive correlations with revenue (0.32 and 0.30).

##### Number of Reviews and Subscribers:
- There is a strong correlation between these two variables (0.65), indicating that courses with more subscribers also tend to receive more reviews.
##### This chart shows that the number of reviews and subscribers on courses are the biggest contributors to revenue.
